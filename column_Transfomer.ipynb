```python
path="/content/drive/MyDrive/نسخة من cereal-kaggle-crawford-modified - sheet 1.csv"
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import set_config
set_config(transform_output="pandas")
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder , OrdinalEncoder
from sklearn.compose import ColumnTransformer
df=pd.read_csv(path)
df.head()
```

```python
X =df.drop(columns=["rating"])
y=df["rating"]
X_train ,X_test, y_train, y_test = train_test_split(X,y,random_state=42)
```

```python
num_col=X_train.select_dtypes(include="number").columns
mean_imputer=SimpleImputer(strategy="mean")
scaler=StandardScaler()
num_pipe=make_pipeline(mean_imputer,scaler)
tuple_num=("number",num_pipe,num_col)
tuple_num
```

```python
ord_col=["shelf"]
ord_imputer=SimpleImputer(strategy="most_frequent")
order=["bottom","middle","top"]
ord_encoder=OrdinalEncoder(categories=[order])
scaler=StandardScaler()
ord_pipe=make_pipeline(ord_imputer,ord_encoder,scaler)
tuple_ord=("ordinal",ord_pipe,ord_col)
tuple_ord
```

```python
cat_col=X_train.select_dtypes("object").drop(columns=ord_col).columns
cat_imputer=SimpleImputer(strategy="constant",fill_value="Missing")
cat_encoder=OneHotEncoder(sparse_output=False,handle_unknown="ignore")
cat_pipe=make_pipeline(cat_imputer,cat_encoder)
tuple_cat=("categorical",cat_pipe,cat_col)
tuple_cat
```

```python
col_tran=ColumnTransformer([tuple_num,tuple_ord,tuple_cat],verbose_feature_names_out=False)
col_tran.fit(X_train)
```

```python
X_train_proccessing=col_tran.transform(X_train)
X_test_proccessing=col_tran.transform(X_test)
X_test_proccessing.head()
```
